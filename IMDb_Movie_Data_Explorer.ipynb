```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("imdb.csv", low_memory=False)

# Show first 5 rows
print("ðŸ”¹ Preview of dataset:")
print(df.head())

# Drop missing values in 'genre' and 'year'
df = df.dropna(subset=['genre', 'year'])

# Convert 'year' to int
df['year'] = df['year'].astype(int)

# Top 10 genres
print("\nðŸ”¹ Top Genres:")
top_genres = df['genre'].value_counts().head(10)
print(top_genres)

# Plot genre distribution
top_genres.plot(kind='bar', title='Top 10 Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()

# Movies released per year (2000â€“2020)
df_filtered = df[df['year'].between(2000, 2020)]
movies_per_year = df_filtered['year'].value_counts().sort_index()

# Plot movies by year
movies_per_year.plot(kind='line', marker='o', title='Movies Released Per Year (2000â€“2020)')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# IMDb Movie Data Explorer
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("imdb.csv", low_memory=False)  # Change filename if different

# View column names
print("ðŸ“Œ Columns in dataset:")
print(df.columns)

# Preview first 5 rows
print("\nðŸ”¹ Preview of dataset:")
print(df.head())

# -----------------------------------------
# GENRE ANALYSIS
# -----------------------------------------
df = df.dropna(subset=['genre'])
top_genres = df['genre'].value_counts().head(10)

# Plot Top Genres
top_genres.plot(kind='bar', title='Top 10 Movie Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()

# -----------------------------------------
# RATING ANALYSIS
# -----------------------------------------
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df = df.dropna(subset=['rating'])

# Plot Rating Distribution
df['rating'].plot(kind='hist', bins=20, title='IMDb Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Number of Movies')
plt.grid(True)
plt.tight_layout()
plt.show()

# -----------------------------------------
# CERTIFICATE ANALYSIS
# -----------------------------------------
df = df.dropna(subset=['certificate'])
top_cert = df['certificate'].value_counts().head(10)

# Plot Certificates
top_cert.plot(kind='bar', title='Top Movie Certificates')
plt.xlabel('Certificate')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# -----------------------------------------
# MOST VOTED MOVIES
# -----------------------------------------
df['votes'] = pd.to_numeric(df['votes'], errors='coerce')
df = df.dropna(subset=['votes'])

top_voted = df.sort_values('votes', ascending=False).head(10)
print("\nðŸŽ¥ Top 10 Most Voted Movies:")
print(top_voted[['lister-item-header', 'votes']])
```
